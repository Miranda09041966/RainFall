```python
import numpy as np
import pandas as pd
from numpy import linalg as LA

def multiple_five(a):
  my_result =(a%5==0)&(a%2==1)
  return a[my_result]

numbers=np.array([5,10,15,22,25,30,31,45,50,55,57])
resp=multiple_five(numbers)
#print(f"multiple of five numbers:{resp}")

def normalize(a):
    # WRITE ME
    soma=0
    for i in range(len(a)):
      soma=soma+a[i]*a[i]
      raiz=np.sqrt(soma)
    return raiz

a=np.array([3,4])
print(normalize(a))
```
